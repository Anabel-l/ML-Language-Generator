```python
vowels = [[ "a", "a", "\u00E1", "\u00E0" ], #A
			[ "e", "e", "\u00E9", "\u00E8","ee"], #E
			[ "o", "o", "\u00F2", "\u00F3","oo"], #O
			[ "u", "u", "\u00FA", "\u00F9" ], #U
			[ "i", "i", "\u00EC", "\u00ED"], #I
			["a'a","aa", "a'a", "ai", "ae", "ao", "au","e'e", "ei", "ea", "eu", "eo","o'o","oa", "oi", "ou", "oe","u'u", "uu","ue", "ua", "ui", "uo","i'i","ia", "ii","ie", "io", "iu"], #doubles
			[ "'","-"]]; #silence
```

```python
consonants = [[ "b", "d", "v", "g"],
			[ "w", "l", "m", "n", "r", "s", "y"],
			[ "c", "z", "s"],
			[ "h", "f", "th", "v", "z", "j"],
			[ "q", "k", "c", "t", "p", "x" ],
			[ "sh", "j", "wh", "f", "s"],
			[ "b", "c", "d", "g", "l", "m", "n", "p", "r", "s", "t"],
			[ "d", "g", "l", "n", "r", "s", "t"],
			[ "st", "dj", "tr", "dr", "sd", "kj", "kr", "cr", "ck", "rx"],
			["mw","nw","mb","nb","ng", "nh", "hn"]];
```

```python
#modules needed
import numpy as np
from pathlib import Path
import re
import json
import sys
import collections
import os
import random
import torch
import torch.nn as nn
```

```python
class WordStruct():
    def __init__(self):
        self.lang_diversity = 3
        self.c = "1"
        self.v = "0"
        self.vowelset = np.empty((len(vowels)) * self.lang_diversity, dtype=object)
        self.conset = np.empty((len(consonants)) * self.lang_diversity, dtype=object)
        self.word_struct = np.empty(self.lang_diversity * 2, dtype=object)

    def fillV(self, i, count):
        self.vowelset[count] = vowels[i][np.random.randint(low=0, high=(len(vowels[i]) - 1))]

    def fillVowels(self, normality):
        count = 0
        for b in range(self.lang_diversity):
            for i, vowset in enumerate(vowels):
                rand = np.random.randint(low=0, high=100)
                if i < 5 and rand < normality:
                    self.vowelset[count] = vowset[0]
                elif (i < 5 and rand >=normality) or (i == 5 and rand < 1) or rand < 5:
                    self.fillV(i, count)
                else:
                    pos = np.random.randint(low=0, high=5)
                    self.fillV(pos, count)

                count = count + 1

    def fillConsonants(self):
        count = 0
        for b in range(self.lang_diversity):
            for c in consonants:
                self.conset[count] = c[np.random.randint(low=0, high=(len(c) - 1))]
                count = count + 1

    def addLetter(self, prob):
        rand = np.random.randint(low=0, high=100)
        if (rand < prob): return self.c
        else: return self.v

    def fillWordStruct(self):
        for i in range(len(self.word_struct)):
            wd = ""
            length = 0
            rand = np.random.randint(low=0, high=100)
            if rand <= 75:
                length = np.random.randint(low=1, high=5)
            elif rand <= 80:
                length = np.random.randint(low=5, high=7)
            elif rand <= 90:
                length = np.random.randint(low=7, high=10)
            else:
                length = 2

            for b in range(length):
                if b == 0 and length == 1:
                    wd = wd + self.addLetter(0)
                elif b == 0:
                    wd = wd + self.addLetter(50)
                elif wd[b - 1: b] == self.c :
                    wd = wd + self.addLetter(5)
                elif wd[b - 1: b] == self.v :
                    wd = wd + self.addLetter(90)

            self.word_struct[i] = wd

    #call this every time you generate a language :)
    def fillAll(self, vow_prob):
        self.fillVowels(vow_prob)
        self.fillConsonants()
        self.fillWordStruct()


    #this section generates words

    def getC(self):
        return self.conset[np.random.randint(low=0, high=(len(self.conset)))]

    def getV(self):
        return self.vowelset[np.random.randint(low=0, high=(len(self.vowelset)))]

    def makeWord(self):
        word = []
        rand = np.random.randint(low=0, high=(len(self.word_struct)))
        for i in range(len(self.word_struct[rand])):
            if self.word_struct[rand][i:i+1] == self.c:
                word.append(self.getC())
            else:
                word.append(self.getV())
        return ''.join(word)
```

```python
#This cell sets up the model.
import re
import json
import sys
import collections
import os
import torch
import torch.nn as nn
from datetime import datetime


# single-direction RNN, optionally tied embeddings
class Emb_RNNLM(nn.Module):
    def __init__(self, params, use_LSTM=True):
        super(Emb_RNNLM, self).__init__()
        self.vocab_size = params['vocab_size']
        self.d_emb = params['d_emb']
        self.n_layers = params['num_layers']
        self.d_hid = params['d_hid']
        self.embeddings = nn.Embedding(self.vocab_size, self.d_emb)
        self.use_LSTM = use_LSTM
        if use_LSTM:
            print('Using LSTM model')
            self.i2R = nn.LSTM(self.d_emb, self.d_hid, batch_first=True, num_layers=self.n_layers) #input to recurrent layer, default nonlinearity is tanh
        else:
            # input to recurrent layer, default nonlinearity is tanh
            self.i2R = nn.RNN(
                self.d_emb, self.d_hid, batch_first=True, num_layers = self.n_layers
            )
        # recurrent to output layer
        self.R2o = nn.Linear(self.d_hid, self.vocab_size)

    def forward(self, train_datum):
        embs = torch.unsqueeze(self.embeddings(train_datum), 0)
        if self.use_LSTM:
            output, (hidden, context) = self.i2R(embs)
        else:
            output, hidden = self.i2R(embs)
        return self.R2o(output) #The second and third returned values are not used for training but for probing the model to see what it is encoding.
```

```python

class Dictionary():
    def __init__(self):
        self.Lang = WordStruct()
        self.Lang.fillAll(95)
        self.word_list = []
        self.vocab = []
        self.characters = []
        self.sentences_as_indices = []
        self.params = {}
        self.wd2ix = {}


    def fillNounList(self):
        print('Processing English words')

            # Open the file in "read" mode ('r')
        with open('drive/MyDrive/Computational Linguistics/Words.txt', 'r') as text:
            for i, line in enumerate(text.readlines()):
                if i > 1000: break
                if i % 100 == 0:
                    print('Processed', i, 'lines.')

                line = line.rstrip()
                if len(line) < 1:
                    continue

                sentence_buffer = []
                word = self.Lang.makeWord()
                while word in self.vocab:
                  word = self.Lang.makeWord()
                sentence_buffer.append(['<s>'] + [line] + [word] + ['<e>'])
                self.vocab.append(word)
                self.word_list.append(sentence_buffer)
                sent = [ch for sublist in list(sentence_buffer) for ch in sublist]
                self.characters += sent
                #print(sentence_buffer)
        print(self.word_list)

    def fillJson(self):

        total_words = len(self.characters)
        print('total characters', total_words)
        for i, word in enumerate(self.characters):
            print(i, word)
            self.wd2ix[word] = i

        with open('drive/MyDrive/Computational Linguistics/Words.wd2ix.json', 'w') as f3:
            json.dump(self.wd2ix, f3)

        self.sentences_as_indices = [torch.LongTensor([self.wd2ix[w] for w in sent])
            for s in self.word_list for sent in s
          ]
        print("Sentences as indices")
        print(self.sentences_as_indices)
        print("Characters")
        print(self.characters)
        self.params = {'vocab_size': total_words, 'd_emb': 128, 'num_layers': 1, 'd_hid': 128, 'lr': 0.0003, 'epochs': 3}
        self.model = Emb_RNNLM(self.params)
```

```python
dict = Dictionary()
dict.fillNounList()
dict.fillJson()
print(dict.characters)
```

```python

criterion = nn.CrossEntropyLoss(ignore_index=0)
optimiser = torch.optim.Adam(dict.model.parameters(), lr=dict.params['lr'])

save_path = 'drive/MyDrive/Computational Linguistics/words.txt_lstm_checkpoint.pth'

for epoch in range(dict.params['epochs']):
    ep_loss = 0
    random.shuffle(dict.sentences_as_indices)
    for j, train_datum in enumerate(dict.sentences_as_indices):
        if len(train_datum) < 4:
            continue

        preds = dict.model(train_datum)

        preds = preds[:, :-1, :].contiguous().view(-1, dict.params['vocab_size'])

        targets = torch.unsqueeze(train_datum, 0)
        targets = targets[:, 1:].contiguous().view(-1)


        loss = criterion(preds, targets)

        if torch.isnan(loss):
            print(train_datum)
        loss.backward()
        optimiser.step()
        optimiser.zero_grad()
        ep_loss += loss.detach()
        if j > 0 and j % 1000 == 0:
            print('processed', j, 'training examples')
    print('Saving checkpoint')
    torch.save({'net_state_dict': dict.model.state_dict(),  'optimiser_state_dict': optimiser.state_dict()}, save_path)
    print('epoch', epoch, 'epoch loss', ep_loss / len(dict.sentences_as_indices))

```
